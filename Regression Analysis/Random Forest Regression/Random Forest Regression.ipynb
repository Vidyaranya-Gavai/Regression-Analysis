```python
# Importing libraries and dataset

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('F:\MachineHack\Regression Analysis\Random Forest Regression\\advertising.csv')
data
```

```python
# Defining input and output features

x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
print(x.shape, y.shape)
```

```python
# defining training and testing dataset

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Defining and training Random forest regressor model

from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(n_estimators = 10)
reg.fit(x_train, y_train)
```

```python
# Making predictions on testing dataset

y_pred = reg.predict(x_test)

# Comparing the predicted sales with actual sales

pd.DataFrame(data={'Predicted': y_pred, 'Actual': y_test})
```

```python
# Evaluating the model perforamace

from sklearn.metrics import mean_squared_error, r2_score
import math

mse = mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('MSE = ', mse)
print('RMSE = ', rmse)
print('R-Squared = ', r2)
```

```python
# As you can see that the value of R-squared is very
# close to 1.
# This indicates that our model is well fitted.
```
