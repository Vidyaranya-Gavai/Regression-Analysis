```python
# Importing libraries and dataset
import numpy as np
import pandas as pd

data = pd.read_csv('F:\MachineHack\Regression Analysis\KNN Regression\Profit.csv')
```

```python
data
```

```python
# Defining input and output features

x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
print(x.shape, y.shape)
```

```python
# Defining training and testing datasets

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Initializing KNN Regression model

from sklearn.neighbors import KNeighborsRegressor

reg = KNeighborsRegressor()

# Finding the optimal value of k (Hyperparameter tuning)

from sklearn.model_selection import GridSearchCV

k_range = list(range(1,21))
param_grid = dict(n_neighbors = k_range)
grid = GridSearchCV(reg, param_grid, cv=10, scoring='r2', return_train_score=False, verbose=4)
grid.fit(x_train, y_train)

# Training the model

reg.fit(x_train, y_train)
```

```python
# Making predictions on testing data

y_pred = reg.predict(x_test)

# Comparing the predicted and actual values of profit

pd.DataFrame(data = {'Predicted':y_pred, 'Actual':y_test})
```

```python
# Evaluating model performance

from sklearn.metrics import mean_squared_error, r2_score
import math

mse = mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print('MSE = ', mse)
print('RMSE = ', rmse)
print('R-Squared = ', r2)
```
